```python
# 2. Pobrać embeddingi dla tych 200k próbek (streaming)
df = pd.read_parquet('mappool_74M_scores_full.parquet')

# Stratyfikuj 200k indexów (bez sprawdzania URLs)
selected_indices = stratified_sample(df, 200_000)

# Pobierz embeddingi dla tych 200k
dataset = load_dataset("sraimund/MapPool", split="train", streaming=True)

embeddings = []
indices = []

for count, row in enumerate(tqdm(dataset, total=75_905_000, desc="Pobieranie embeddingów", unit=" rekordów")):
    if count in selected_indices:
        if 'embedding' in row and row['embedding'] is not None:
            embeddings.append(row['embedding'])
            indices.append(count)

        if len(indices) == len(selected_indices):
            break

np.savez_compressed('AL_embeddings_200k.npz',
                    embeddings=embeddings,
                    indices=indices)
```
