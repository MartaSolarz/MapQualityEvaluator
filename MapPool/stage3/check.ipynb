```python
# read jsonl file and save to new file this whoch label = YES or NO
import json

with open('./iterations/iteration_2/annotations_progress.jsonl', 'r') as f:
    data = [json.loads(line) for line in f]
new_data = [item for item in data if item['label'] in ['YES', 'NO']]

with open('train_yes_no.jsonl', 'w') as f:
    for item in new_data:
        json.dump(item, f)
        f.write('\n')
```
